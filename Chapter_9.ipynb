####**Strategic Management**
The ***strategic management*** process is the set of activities that firm managers undertake in order to try to put their firms in the best
possible position to compete successfully in the marketplace.


---






####**Strategic Cycle**

Strategic management is made up of several
distinct activities (**S DC I M V**) that is 


**S**tategic Analysis $\implies$
**D**evelop Objectives $\implies$
**C**reate & Choose Strategies $\implies$
**I**mplement Strategies $\implies$
**M**easure & Evaluate Performance $\implies$
**V**ision & Mission 



> Most top managers deal with all of the steps simultaneously.

> They engage in environmental scanning to update their analytical view of
the firm, they are executing strategies formulated in the past, they are formulating strategies to execute in the

####**Vision Statement**
> A **vision statement** is an expression of what a business’s founders want that business to accomplish. 

The **vision statement** is usually very broad, and it **does not even have to mention a product or service.**

The vision statement **does not describe the strategy** a firm will use to follow its vision—it is **simply a sentence or two that states why the business exists**.



####**Mission Statement**

> **Mission Statement** a general description of how the firm will try to accomplish the firm’s vision.



The **mission statement** takes the <ins>why of a vision statement</ins> and **gives a broad description of how the firm will try to make its vision a reality**. 


A mission statement is <ins>still not exactly a strategy</ins>

A mission statement **focuses on describing the products** a firm plans to offer or the **target markets it plans to serve**.





image.png

While a firm’s **vision statement is a general statement** about its values, a **firm’s mission statement is more specific**.

####**The Planning Process (S DIR)**

> The **planning process** is an integral part of the strategy cycle, since
developing objectives, creating strategies, and implementing firm-wide activities all require extensive
planning.

image.png

> When designing a plan, a manager may think of **many ways to achieve a goal**. He or she can have a <ins>group of employees brainstorm to come up with ideas</ins>. **Not all** of the potential ideas are likely to be **feasible**.

> Part of the manager’s task in designing a plan is to **coordinate** various **ideas** with a **firm’s resources and capabilities and its time constraints**

####**Implementing Plans for Different Levels of Firm Activity and Time Horizons**

> **Short-term strategic plan** is one that can be accomplished in a year or sooner. 

> A **Long-term strategic plan** is developed when an objective cannot be accomplished in less than a year.




####**Scale Levels of Planning**

> The day-to-day planning that managers do to take steps toward those bigger objectives is key to achieving success.

#####**Strategic planning** 

The high-level planning performed by **company executives** in order to set the overall direction of the company. Grand
strategies are part of strategic planning, as are business-level strategies such as cost leadership. 

>Strategic planning connects the company’s actions back to its vision and mission statements (the “why does this company exist” question).

#####**Tactical planning** 

Tactical planning is **mid-level planning** that consists of broad ideas of what the company should do to **pursue its mission.** This is the sort of planning done by **division managers**.

#####**Operational planning**

Operational planning lays out the **front-line** activities that each employee in the company will do to advance the tactical plans.

Operational plans are the **daily activities** required for the company to function, including <ins>ordering inventory or supplies, scheduling workers and defining their work tasks, and developing sales goals and promotions to help achieve those goals.</ins>

####**Implementing Planned Strategies**
Implementation of planned strategies refers to the e**xecution of a strategy by assigning tasks for people** to carry out to accomplish the company’s strategic goals.

####**Measuring and Evaluating Strategic Performance**

Performance measurement comes in many forms, from financial reports to quality measures like defect
rates. 

> **Performance measurement:** Any activity a firm can perform can have a performance measure developed to evaluate the success of
that activity.

Evaluation involves setting a <ins>*performance standard, measuring the results of firm activities, and comparing the results to the standard.*</ins>

> **Benchmarking**, a process in which the performance standard is based on another firm’s superior performance.

image.png

The strategy cycle is a process managers use to achieve an <ins>advantage in the marketplace, and the measurement and evaluation stage</ins> tells managers whether the advantage is being achieved. If firm performance meets or exceeds objectives, then the manager reports the success to **middle and upper-level managers**.


> **Performance feedback** becomes part of the strategic analysis of the
**firm’s capabilities and resources, and firm leadership** uses the information to help develop better strategies for firm success.
